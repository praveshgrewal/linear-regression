linear regression

```python
import pandas as pd
hours = [2,3,4,5,6,7,]
marks= [43,50,55,60,65,68]
```

plot

```python
import matplotlib.pyplot as plt
plt.scatter(hours,marks)
```

```python
#hear map
df= pd.DataFrame(hours,marks)
import seaborn as sns
sns.heatmap(df.corr())
```

```python
#regression plot
sns.regplot(x= hours, y = marks)
```

```python
import sklearn.linear_model
#algorithm store
regressor = sklearn.linear_model.LinearRegression()
#train algorithm

regressor.fit(hours,marks)
```

```python
#test our data
x_test= [[8],[9]]
y_pred = regressor.predict(x_test)
```
